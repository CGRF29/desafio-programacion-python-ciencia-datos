Extraer solamente aquellas columnas que sean de tipo float o int

```python
#Libreria
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('empleados.csv',delimiter=";")
df.head()
```

```python
df.info()
```

```python
#extraccion_df = df.loc[:, ['Edad', 'Experiencia', 'Salario']]
#extraccion_df.head()
columnas_numericas = df.select_dtypes(include=['float','int'])
columnas_numericas
```

Extraer filas que cumplan con un criterio

```python
df_ = np.array(df)
```

```python
edad_mayor_30 = df_[df_[:,4] > 29.9]
```

```python
edad_mayor_30
```

Extraer los registros más altos

```python
first_rows = df_[df_[:,4].argsort()[::-1]][:10]
first_rows
```
