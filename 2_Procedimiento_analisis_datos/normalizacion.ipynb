##### Normalización: Ajusta los valores de las diferentes columnas para que existan en una escala común, lo que permite compararlos más fácilmente

Actividad: Aplicar una técnica de normalización en las columnas del dataframe

```python
#Librerias
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('datos.csv', delimiter = ";")
df.head()
```

```python
scaler = MinMaxScaler()
```

```python
scaler_df =scaler.fit_transform(df)
```

```python
scaler_df = pd.DataFrame(scaler_df, columns = ['Altura','Peso'])
```
