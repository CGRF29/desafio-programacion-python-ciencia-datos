```python
# Library
import pandas as pd
```

```python
df = pd.read_csv('empleados.csv',delimiter=';')
print(df.head())
```

##### Identificación de valores faltantes

```python
df.isnull().sum()
```

```python
print(df.notnull().sum())
```

```python
df[df.isnull().any(axis=1)]
```

```python
df_ = df.drop(axis=1, columns=['Apellido'])
```

```python
df_
```

```python
df_actualizado = df_.dropna(axis=0)
df_actualizado
```

```python
df_actualizado_ = df_.dropna(axis=0, thresh =5)
df_actualizado_
```

```python
duplicados = df_actualizado.duplicated()
print(df_actualizado[duplicados])
```
